#Data Dictionary
##Variable Description

Survived - Survived (1) or died (0)
Pclass - Passenger’s class (1 = 1st, 2 = 2nd, 3 = 3rd)
Name - Passenger’s name
Sex - Passenger’s sex
Age - Passenger’s age
SibSp - Number of siblings/spouses aboard
Parch - Number of parents/children aboard (Some children travelled only with a nanny, therefore parch=0 for them.)
Ticket - Ticket number
Fare - Fare
Cabin - Cabin
Embarked - Port of embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)

```python
def getNumber(str):
    if str=="male":
        return 1
    else:
        return 2
data["gender"]=data["Sex"].apply(getNumber)
#We have created a new column called "gender" and 
#filling it with values 1 ,2 based on the values of sex column
data.head()
```

```python
#Deleting Sex column, since no use of it now
del data["Sex"]
data.head()
```

```python
data.isnull().sum()
```

```python
#finding mean survived age
meanS= data[data.Survived==1].Age.mean()
meanS
```

```python
data["age"]=np.where(pd.isnull(data.Age) & data["Survived"]==1  ,meanS, data["Age"])
data.head()
```

```python
# Finding the mean age of "Not Survived" people
meanNS=data[data.Survived==0].Age.mean()
meanNS
```

```python
data.age.fillna(meanNS,inplace=True)
data.head()
```

```python
#We can safely delete the Age column
del data["Age"]
data.head()
```

```python
#Renaming "age" and "gender" columns
data.rename(columns={'age':'Age'}, inplace=True)
data.rename(columns={'gender':'Sex'}, inplace=True)
data.head()
```
