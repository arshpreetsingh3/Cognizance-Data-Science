```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADJ
    else:
             return wordnet.NOUN
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
stops,string.punctuation
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w,pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
documents = [(clean_review(document),category) for document,category in documents]
```

```python
training_documents = documents[0:75]
testing_documents = documents[75:]
```

```python
all_words = []
for doc in training_documents:
    all_words += doc[0]
    #because doc is a tuple where 0 index refers to all tthe words
```

```python
import nltk 
```

```python
freq = nltk.FreqDist(all_words)
freq.most_common(3000)
```

```python
#now we dont require the word frequency we want to know thw word instead 
```

```python
common = freq.most_common(3000)
features = [i[0] for i in common ]
```

```python
features
```

```python
#thus these are the 3000 words that wold be our features
```

```python
#now we want to create a dictionary foreach document where dictionary will have the following format 
#eg= film : true/false that is whether film word is present in the document or not 
```

```python
documents[0]
#this contains all the words which are there in this soecific review 
```

```python
def get_feature_dict(words):
    current_features = {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
        #means if w comes in word set then it'll be true otherwise false
    return current_features
        
    
```

```python
get_feature_dict(training_documents[0][0])
```

```python
training_data = [(get_feature_dict(doc),category)for doc,category in training_documents]
testing_data = [(get_feature_dict(doc),category)for doc,category in testing_documents]
```

```python
training_data[0]
```

```python
#nltk has inbuilt naive bayes classifier that we'll use
#nltk classifier require data in this form 
```

```python
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier,testing_data)
#this has the accracy function
```
