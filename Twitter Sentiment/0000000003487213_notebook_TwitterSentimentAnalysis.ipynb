```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
```

```python
stop=set(stopwords.words('english'))
airline_names = ['@united', '@usairways', '@americanair', '@southwestair', '@jetblue', '@virginamerica']
punctuation=list(string.punctuation)
stop.update(punctuation)
stop.update(airline_names)
```

```python
stop
```

```python
from nltk import pos_tag
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
def clean_review(words):
    output_words=[]
    words=words.split()
    for w in words:
        if w.lower() not in stop and len(w.lower())>1:
            pos=pos_tag([w])[0][1]
            clean_word=lemmatizer.lemmatize(w,pos=get_simple_pos(pos))
            output_words.append(clean_word.lower())
    return output_words 
```

```python
def clean(documents):
    new_document=[]
    for words in documents:
        clean_word=clean_review(words)
        new_document.append((clean_word))
    return new_document    
```

```python
X_text_train_clean=clean(X_text_train)
```

```python
X_text_test_clean=clean(X_test_test)
```

```python
X_text_train_sent=[" ".join(document) for document in X_text_train_clean]
```

```python
X_text_test_sent=[" ".join(document) for document in X_text_test_clean]
```
