```python
import numpy as np
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
train_data = pd.read_csv('0000000000002747_training_twitter_x_y_train.csv',delimiter=',')
test_data = pd.read_csv('0000000000002747_test_twitter_x_test.csv',delimiter = ',')
train_target = train_data['airline_sentiment']
#train_data.drop(1,inplace = True)
train_target
pos_text = train_data[train_data.airline_sentiment == 'positive'].text
neg_text = train_data[train_data.airline_sentiment == 'negative'].text
neutral_text = train_data[train_data.airline_sentiment == 'neutral'].text
#train_tweet = train_data['text']
test_tweet = test_data['text']
categories = set(train_target)
#documents = []
#for category in categories:
#    for tweet in train_tweet:
documents = []
for tweet in pos_text:
        documents.append((word_tokenize(tweet),'positive'))
for tweet in neg_text:
        documents.append((word_tokenize(tweet),'negative'))
for tweet in neutral_text:
        documents.append((word_tokenize(tweet),'neutral'))        
```

```python
documents[0:5]
import random
random.shuffle(documents)
```

```python
docu = []
for tweet in test_tweet:
    docu.append(word_tokenize(tweet))
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk import pos_tag
from nltk.corpus import wordnet
def get_simple_pos(tag): 
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
#stops, string.punctuation
```

```python
def clean_review(words):
    output_words = []
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
train_documents = [(clean_review(document),category) for document,category in documents]
```

```python
train_documents
```

```python
test_docu = [clean_review(document) for document in docu]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
x_train = [" ".join(document) for document,category in train_documents]
```

```python
x_test =  [" ".join(document) for document in test_docu]
```

```python
count_vec = CountVectorizer(max_features = 3000)
x_train_features = count_vec.fit_transform(x_train)
x_train_features.todense()
```

```python
x_test_features = count_vec.transform(x_test)
```
