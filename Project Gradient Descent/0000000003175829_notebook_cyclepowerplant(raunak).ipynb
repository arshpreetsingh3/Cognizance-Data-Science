```python
import numpy as np
from sklearn.metrics import r2_score
from sklearn import preprocessing
```

```python
def predict(X,m,k):
    M=len(X)
    Y_test=[0 for i in range(0,M)]
    for i in range(0,M):
        sum=0
        for j in range(0,k): #
            sum+=X[i][j]*m[j]
        Y_test[i]=sum
    return Y_test
```

```python
def step_gradient(X,Y,learning_rate,m,k):
    m_slope=[0 for i in range(0,k)]
    new_m=[0 for i in range(0,k)]
    N=len(X)
    for i in range(0,N):
        sum=0;
        for j in range(0,k):
            sum+=(m[j]*X[i][j])
        for j in range(0,k):
            m_slope[j]+=(-2/N)*(Y[i]-sum)*X[i][j]
    for k in range(0,k):
        new_m[k]=m[k]-learning_rate*m_slope[k]
    return new_m
```

```python
def cost(X,Y,m,k):
    total=0
    M=len(X)
    for i in range(0,M):
        sum=0
        for j in range(0,k):
            sum+=m[j]*X[i][j]
        total+=1/M*((Y[i]-sum)**2)
    return total
```

```python
def gd(X,Y,learning_rate,num_iterations,k):
    m=[0 for i in range(0,k)]
    for i in range(0,num_iterations):
        m=step_gradient(X,Y,learning_rate,m,k)
        cos=cost(X,Y,m,k)
        print(cos)
    return m
```
