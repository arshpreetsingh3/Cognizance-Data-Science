```python
import numpy as np
import pandas as pd
```

```python
def step_gradient(points,learning_rate,m,c):
    m_slope=[0.0 for i in range(len(points[0])-1)]
    c_slope=0
    m_new=[0.0 for i in range(len(points[0])-1)]
    c_new=0
    n=len(points)
    col = len(points[0])-1
    for j in range(col):
        for i in range(len(points)):
            x = points[i,j]
            y = points[i,-1]
            m_slope[j] += (-2/n)* (y - m[j] * x - c)*x
            c_slope += (-2/n)* (y - m[j] * x - c)
        m_new[j] = m[j] - (learning_rate*m_slope[j])
        c_new = c - (learning_rate*c_slope)
    return m_new, c_new   
```

```python
def gd(data,learning_rate,steps):
    m=list([0 for i in range(len(data[0])-1)]) 
    c=0
    for i in range(steps):
        m,c = step_gradient(data,learning_rate,m,c)
        #print("Trail ",i)
        #print(m)
        #print(c)
    return m,c
```

```python
def predict(m,c):
    ypred=0
    train = np.loadtxt(r"C:\Users\212688702\Desktop\ML\boston_test.csv",delimiter=",")
    for j in range(len(train)):
        ypred=0
        for i in range(len(train[0])-1):
            ypred+=train[j,i]*m[i]
        ypred+=c
        with open('somefile.csv', 'a') as the_file:
            the_file.write(str(ypred)+"\n")
```

```python
def run():
    data = np.loadtxt(r"C:\Users\212688702\Desktop\ML\boston_train.csv",delimiter=",")
    learning_rate = 0.0001
    steps = 1000
    m,c = gd(data,learning_rate,steps)
    #print(m)
    #print(c)
    predict(m,c)
```
